```python
import matplotlib.pyplot as plt

# Number of processes (P)
processes = [2, 4, 8, 16]

# Execution times for each method (in seconds)
execution_times_jacobi = [5.807598352432251, 2.047309398651123, 0.7131476402282715, 0.28354549407958984]
execution_times_gauss_seidel = [6.69243860244751, 2.136303186416626, 0.7217752933502197, 0.29558491706848145]

# Baseline time (P=1 is estimated as the time with 2 processes, for simplicity)
baseline_time_jacobi = execution_times_jacobi[0]
baseline_time_gauss_seidel = execution_times_gauss_seidel[0]

# Calculate speed-up
speedup_jacobi = [baseline_time_jacobi / time for time in execution_times_jacobi]
speedup_gauss_seidel = [baseline_time_gauss_seidel / time for time in execution_times_gauss_seidel]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(processes, speedup_jacobi, marker='o', label='Jacobi Method')
plt.plot(processes, speedup_gauss_seidel, marker='s', label='Gauss-Seidel Method')

plt.xlabel("Number of Processes (P)")
plt.ylabel("Speed-up")
plt.title("Speed-up vs. Number of Processes for Jacobi and Gauss-Seidel Methods")
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```
